# Plot customization

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,15,25,30]
plt.plot(
    x,
    y,
    color='green',
    linestyle='--',
    marker='o'
)
plt.title('My Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

# Multiple Lines

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]

y1 = [1, 4, 9]
y2 = [1, 2, 3]

plt.plot(x, y1, label='Squares')
plt.plot(x, y2, label='Linear')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3]

y1 = [1,4,9]
y2 = [1,2,3]

plt.plot(
    x,
    y1,
    color='red',
    marker='o',
    label='Squares'
)

plt.plot(
    x,
    y2,
    color='blue',
    marker='s',
    label='Linear'
)

plt.title('Comparison Graph')
plt.xlabel('X Values')
plt.ylabel('Y Values')

plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values,width=0.5,color=['green','red','blue'])
plt.grid(True)
plt.title("Category Comparison")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```

# Histogram

```python
#plt.hist(data)
```

```python
import matplotlib.pyplot as plt

data = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.hist(data, bins=5, color='green',edgecolor='black')

plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")


plt.show()
```

# Scatter plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y,color=['red','green','blue','orange'],s=100,marker='*')
plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)



plt.show()
```

# Subplots

```python
import matplotlib.pyplot as plt


plt.subplot(1,2,1)
plt.plot([1,2,3],[1,2,3])
plt.title("Increasing")

plt.subplot(1,2,2)
plt.plot([1,2,3],[3,2,1])
plt.title("Decreasing")


plt.show()
```
