# Filtering Data

```python
# Age greater than 25
```

```python
df[df['Age']>25]
```

```python
# Marks greater than 88
```

```python
df[df['Marks']>88]
```

```python
#Age equal to 30
```

```python
df[df['Age']==30]
```

```python
# Age not equal to 25
```

```python
df[df['Age']!=25]
```

```python
df[(df['Age'] > 25) & (df['Marks'] > 85)]
```

```python
df[(df['Age'] > 25) | (df['Marks'] > 90)]
```

```python
#Students with Age > 25, show only Name:
```

```python
df[df['Age'] > 25]['Name']
```

# Grouping and aggregating

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Gender": ["F", "M", "M", "F"],
    "Age": [25, 30, 22, 28]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.groupby('Gender')['Age'].mean()
```

```python
df.groupby('Gender')['Age'].sum()
```

```python
df.groupby('Gender')['Age'].count()
```

```python
df.groupby('Gender')['Age'].max()
```

```python
df.groupby('Gender')['Age'].min()
```

```python
df.groupby('Gender')['Age'].agg(['mean','max','min','count'])
```

# Merging and Joining

```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Marks': [90, 85, 88]
})
```

```python
pd.merge(df1, df2, on='ID')
```

```python
#on='ID'
#Match rows where ID is the same.
```

```python
df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob']
})

df2 = pd.DataFrame({
    'Marks': [90, 85]
})
```

```python
df1.join(df2)
```
